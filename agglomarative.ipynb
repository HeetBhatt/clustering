```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv")
df.head()
```

```python
df.head()
```

```python
import scipy.cluster.hierarchy as sch
```

```python
df.shape
```

```python
data = df.iloc[:,3:5].values
```

```python
plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = sch.dendrogram(sch.linkage(data, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
labels = agg.fit_predict(data)  
```

```python
labels
```

```python
plt.figure(figsize =(10,5))
plt.scatter(data[:,0],data[:,1],cmap="rainbow",c=agg.labels_)
```
